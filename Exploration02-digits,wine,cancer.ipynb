#  Load_digits 손글씨 분류하기

손글씨는 글씨가 맞는지 아닌지를 판별하는게 기준이라서 평가지표로는 Accuracy로 결정

```python
#필요한 모듈
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression



#1. 데이터 준비하기
digits = load_digits()
print(dir(digits))   #어떤변수랑 메서드 가지고있는지 확인
digits_data = digits.data
digits_label = digits.target
digits.target_names  #Target Names 출력해 보기
print(digits.DESCR)  #데이터 Describe 해 보기


#2. train, test데이터 분리하기
X_train, X_test, y_train, y_test = train_test_split(digits_data, digits_label, test_size=0.2, random_state=7)
```

```python
#3. 여러가지 모델 학습 및 예측해보기
# DecisionTree로 분류해보기 
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred),'Accuracy: ', accuracy)


#Random Forest 사용 
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred),'Accuracy: ', accuracy)


#svm모델 사용
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred),'Accuracy: ', accuracy)


#SGD Classifier 사용 
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred),'Accuracy: ', accuracy)


#Logistic Regression 사용 - 0.978
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(classification_report(y_test, y_pred),'Accuracy: ', accuracy)

```

# load_breast_cancer - 유방암 여부를 진단하기

암 환자인데 암 환자가 아니라고 하는 경우를 최대한 줄이는게 좋기때문에 recall이 중요함 - recall_score사용

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score, accuracy_score
from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

#1. 데이터 준비하기
breast_cancer = load_breast_cancer()
breast_cancer_data = breast_cancer.data
breast_cancer_label = breast_cancer.target
#유방암 데이터 feature
print(breast_cancer.feature_names)
print(breast_cancer.target_names)
print(breast_cancer.DESCR)

#2. train, test 데이터 분리하기
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, breast_cancer_label, test_size=0.2, random_state=7)

```

```python
#3. 모델 학습 및 예측
#3. 여러가지 모델 학습 및 예측해보기
#DecisionTree로 분류해보기 
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)
print(classification_report(y_test, y_pred))

recall = recall_score(y_test, y_pred, average='weighted')
print('recall_score : ', recall)


#Random Forest 사용 
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
print(classification_report(y_test, y_pred))

recall = recall_score(y_test, y_pred, average='weighted')
print('recall_score : ', recall)


#svm모델 사용
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print(classification_report(y_test, y_pred))

recall = recall_score(y_test, y_pred, average='weighted')
print('recall_score : ', recall)


#SGD Classifier 사용 
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
print(classification_report(y_test, y_pred))

recall = recall_score(y_test, y_pred, average='weighted')
print('recall_score : ', recall)

#Logistic Regression 사용 
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
print(classification_report(y_test, y_pred))

recall = recall_score(y_test, y_pred, average='weighted')
print('recall_score : ', recall)
```
